2) Crie um programa em Python para adicionar membros a um Set. Adicione a palavra "Red". em seguida faça um update adicionando ["Blue", "Green"]. Print o set ao final. 

````python
set_inicial = set(["Blue", "Green"])
palavra = input("Digite uma palavra para adicionar no set")
set_inicial.add(palavra)
print(set_inicial)
````

3) Crie um programa em Python para remover itens do set. Remova ao menos 2 itens. 

Input:

```
[0, 1, 3, 4, 5]
```

output:
```
{1, 3, 4, 5}        
{3, 4, 5} 
```









````python
definindo_set = set([0, 1, 3, 4, 5])
print(definindo_set)
print("Removendo dois itens")
for i in range(2):
    definindo_set.discard(i)
    print(definindo_set)
````

4) Crie um programa para limpar um Set. Obs.: Copie o conteúdo antes para outro Set.

````python
set_inicial = set(["Blue", "Green"])
set_copiado = set_inicial.copy()

print("set inicial: ", set_inicial)
set_inicial.clear()
print("O set inicial foi limpo ", set_inicial)
print("O set copiado  ", set_copiado)
````

5) Crie um programa em Python para criar a união de dois Sets.
Input:

```
["green", "blue"]
["blue", "yellow"]
```
output:

```
{'yellow', 'green', 'blue'} 
```





````python
set_1 = set(["green", "blue"])
set_2 = set(["blue", "yellow"])

print(set_1 | set_2)
````
