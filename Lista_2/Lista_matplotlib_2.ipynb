8) Calcule os dados de vendas totais para cada produto e mostre-os usando um gráfico de pizza

```python
import matplotlib.pyplot as plt
import numpy as np

arquivo = open("company_sales_data.csv", "r")
numero_mes = []
facecream = []
facewash = []
toothpaste = []
bathingsoap = []
shampoo = []
moisturizer = []
total_units = []
nomes =[]
x = False
nomes = arquivo.readline().strip().split(",")
nomes = nomes[1:-2]
for i in arquivo:
    if(x):
        i = i.split(",")
        numero_mes.append(i[0])
        facecream.append(int(i[1].strip()))
        facewash.append(int(i[2].strip()))
        toothpaste.append(int(i[3].strip()))
        bathingsoap.append(int(i[4].strip()))
        shampoo.append(int(i[5].strip()))
        moisturizer.append(int(i[6].strip()))
    x = True

total_facecream = np.sum(facecream)
total_facewash = np.sum(facewash)
total_toothpaste = np.sum(toothpaste)
total_bathingsoap = np.sum(bathingsoap)
total_shampoo = np.sum(shampoo)
total_moisturizer = np.sum(moisturizer)
total_geral = total_facecream + total_facewash + total_toothpaste + total_bathingsoap + total_shampoo + total_moisturizer
porcentagem = [total_facecream/total_geral, total_facewash/total_geral, total_toothpaste/total_geral,  total_bathingsoap/total_geral, total_shampoo/total_geral, total_moisturizer/total_geral]
plt.pie(porcentagem, labels=nomes, autopct='%1.1f%%')
plt.legend(nomes, loc="lower right", bbox_to_anchor=(1.3, 0))
plt.show()
```

9) Leia o sabonete de banho (Bathing soap) e limpeza facial (facewash) de todos os meses e exiba-o usando o Subplot

```python
import matplotlib.pyplot as plt

arquivo = open("company_sales_data.csv", "r")

numero_mes = []
facewash = []
bathingsoap = []

x = False
for i in arquivo:
    if(x):
        i = i.split(",")
        numero_mes.append(i[0])
        facewash.append(int(i[2].strip()))
        bathingsoap.append(int(i[4].strip()))
    x = True

fig, axs = plt.subplots(2, sharex=True)

axs[0].plot(numero_mes, bathingsoap,  linewidth ="3", color = "black")
axs[0].plot(numero_mes, bathingsoap,'o', color="black")
axs[0].set_title("Sales data of a Bathingsoap")

axs[1].plot(numero_mes, facewash,  linewidth ="3", color = "red")
axs[1].plot(numero_mes, facewash,'o', color="red")
axs[1].set_title("Sales data of a facewash")

plt.ylabel("Sales units in number")
plt.xlabel("Month Number")
plt.show()
```
