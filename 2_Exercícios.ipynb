2)

Seu script será usado no computador de bordo do carro.

input: Quantidade de dinheiro

output: Quantidade de litros possível de comprar com a quantidade de dinheiro e  autonomia em km com este combustível

4) Elabore um programa que calcule o IMC. 

Obs.:
- Utilize a função input() para coletar os dados de altura e peso. 
- Utilize os operadores relacionais.

Considere: 

*   Muito abaixo do peso  - IMC < 17
*   Abaixo do peso normal - 17 <= IMC < 18.5 
*   Peso dentro do normal - 18.5 <= IMC < 25 
*   Acima do peso normal  - 25 <= IMC < 30  
*   Muito acima do peso   - IMC >= 30
  
Exemplo:
  
input:

```
altura = 1.70
peso   = 65.3
```
output:

```
O IMC =  22.5951557093
Muito abaixo do peso  :  False
Abaixo do peso normal :  False
Peso dentro do normal :  True
Acima do peso normal  :  False
Muito acima do peso   :  False
```



````python
altura = float(input("Digite a altura:"))
peso = float(input("Digite o peso:"))

imc = peso / (altura**2)

if(imc < 17):
    muito_abaixo_do_peso = True
else: 
    muito_abaixo_do_peso = False

if(17 <= imc < 18.5):
    abaixo_do_peso_normal = True
else: 
    abaixo_do_peso_normal = False

if(18.5 <= imc < 25):
    peso_dentro_do_normal = True
else: 
    peso_dentro_do_normal = False

if(25 <= imc < 30):
    acima_do_peso_normal  = True
else: 
    acima_do_peso_normal = False

if(imc >= 30):
    muito_acima_peso = True
else: 
    muito_acima_peso = False

print("Muito abaixo do peso : %s"  %muito_abaixo_do_peso )
print("Abaixo do peso normal :  %s" %abaixo_do_peso_normal)
print("Peso dentro do normal : %s" %peso_dentro_do_normal)
print("Acima do peso normal  : %s" %acima_do_peso_normal)
print("Muito acima do peso   : %s" %muito_acima_peso)
````

5) Faça um programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:

*   salário bruto.
*   quanto pagou de IR
*   quanto pagou ao INSS.
*   quanto pagou ao sindicato.
*   o salário líquido. (Salário bruto - descontos = Salário líquido)



````python
ganho_por_hora = float(input("Quanto você ganha por hora?"))
horas_por_mes = float(input("Quantidade de horas trabalhadas no mês?"))

salario = ganho_por_hora*horas_por_mes
IR = salario * (11/100)
INSS = salario * (8/100)
sindicato = salario * (5/100)
salario_liquido = salario - IR - INSS - sindicato

print("salário bruto: %.2f" %salario)
print("Quanto pagou de IR: %.2f" %IR)
print("quanto pagou ao INSS: %.2f" %INSS)
print("Quanto pagou ao sindicato: %.2f" %sindicato)
print("O salário líquido: %.2f" %salario_liquido)
````
