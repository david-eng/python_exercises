1) Escreva um programa em Python para ler um arquivo linha a linha e armazena-las em uma lista. Obs.: Crie uma função.

```python
lista = []


def ler_arquivo(local):
    global lista
    arquivo = open(local, "r")
    for linha in arquivo:
        lista.append(linha)
    arquivo.close()


ler_arquivo("lista9.txt")
print(lista)
```

```python
def ler_arquivo(local):
    arquivo = open(local, "r")
    lido = arquivo.read()
    arquivo.close()
    return lido


def escrever_em_arquivo(local, frase):
    arquivo = open(local, "w")
    arquivo.write(frase)
    arquivo.close()


print("Arquivo incial: ", ler_arquivo("novo_mundo.txt"))
escrever_em_arquivo("novo_mundo.txt", "Seja bem-vindo")
print("Saída no arquivo: ", ler_arquivo("novo_mundo.txt"))
```

4) Escreva um programa em Python para copiar o conteúúdo de um arquivo para outro arquivo.

```python
def ler_arquivo(local):
    arquivo = open(local, "r")
    lido = arquivo.read()
    arquivo.close()
    return lido


def escrever_em_arquivo(local, frase):
    arquivo = open(local, "w")
    arquivo.write(frase)
    arquivo.close()


def copiar_arquivo(inicial, destino):
    escrever_em_arquivo(destino, ler_arquivo(inicial))


print("Arquivo incial: ", ler_arquivo("inicial.txt"))
print("Arquivo Destino: ", ler_arquivo("destino.txt"))
copiar_arquivo("inicial.txt", "destino.txt")
print("Arquivo Destino atualizado: ", ler_arquivo("destino.txt"))
```

 5) Faça um programa contendo uma função que recebe como
argumentos os nomes de dois arquivos. O primeiro arquivo
contém nomes de alunos e o segundo arquivo contém as
notas dos alunos. No primeiro arquivo, cada linha
corresponde ao nome de um aluno e no segundo arquivo,
cada linha corresponde às notas dos alunos (uma ou mais).
Assuma que as notas foram armazenadas como strings, e
estão separadas umas das outras por espaços em branco.
Leia os dois arquivos e gere um terceiro arquivo que contém o
nome do aluno seguido da média de suas notas.

```python
nomes = []
notas = []
media = []


def ler_saida(local):
    arquivo = open(local, "r")
    texto = arquivo.read()
    arquivo.close()
    return texto


def ler_arquivo(local, vetor):
    arquivo = open(local, "r")
    for linha in arquivo:
        vetor.append(linha.rstrip())
    arquivo.close()
    return arquivo


def escrever_em_arquivo(local, texto):
    arquivo = open(local, "w")
    arquivo.write(texto)
    arquivo.close()


def ler_arquivos(arquivo1, arquivo2):
    global nomes, notas
    ler_arquivo(arquivo1, nomes)
    ler_arquivo(arquivo2, notas)


ler_arquivos("nome_alunos.txt", "notas_alunos.txt")

for i in notas:
    i = i.split(" ")
    soma = 0
    for j in range(len(i)):
        soma += int(i[j])
    media.append(round((soma / len(i)), 2))

texto = ""
for i in range(len(nomes)):
    texto += "Aluno: " + str(nomes[i]) + " Media: " + str(media[i]) + "\n"

escrever_em_arquivo("media_alunos.txt", texto)
ler_saida("media_alunos.txt")
```
