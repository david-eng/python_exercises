2) Escreva uma função que verifica se um numero n está dentro do range -100 a 100. Verifique também se o número é inteiro.

Exemplos: 
```
input: testa_numero(10)
saída: 10 é inteiro e está dentro do range.

input: testa_numero(-200)
saída: -200 é inteiro e está fora do range.
```



````python
def testa_numero(n):
    n = float(n)
    if( -100 < n < 100 and n == int(n) ):
        print("%s é inteiro e está dentro do range." %n)
    elif( -100 < n < 100 and n == float(n) ):
         print("%s não é inteiro e está dentro do range." %n)
    elif( n == float(0)):
        print("%s não é inteiro e está fora do range." %n)
    else:
        print("%s é inteiro e está fora do range." %n)

numero = input("Entrada")
testa_numero(numero)
````

3) Faça um programa que converta da notação de 24 horas para a notação de 12 horas. Por exemplo, o programa deve converter 14:25 em 2:25 P.M; 6:44 em 6:44 A.M. A entrada é dada em dois inteiros. O programa deve ler várias entradas e chamar uma função para convertê-las e em seguida imprimir a saída.

````python
def converte_hora(hora):
    if(int(hora[0]) > 12):
        hora_saida = int(hora[0]) - 12
        print("%s:%s P.M" %(hora_saida, hora[1]))
    else:
        hora_saida = int(hora[0])
        print("%s:%s A.M" %(hora_saida, hora[1]))

hora = str(input("Digite a hora: ")) #conforme pedido entrada do tipo (6:44)(H:M)
hora = hora.split(":")
converte_hora(hora)

````


4) Escreva uma função para printar todos os números primos entre 0 e limite, onde limite é o parametro da função.

````python
def verifica_primo(numero):
    divisores = 0
    if(numero == 1):
        return False
    for i in range(2, numero):
        if(i != numero and numero % i == 0):
            divisores += 1
    if(divisores > 0 ):
        return False
    else:
        return True

limite = int(input("Digite o limite: "))
for i in range(1, limite+1, 1):
    if(verifica_primo(i)):
        print(i)
````

5) Elabore uma função em Python onde se possa passar um número arbitrário de parâmetros. Imprima os parametros passado na função.

Exemplo:

input:


```
imprime_parametros('python','e','muito','legal')
imprime_parametros(1,2,3,3,2,1)
```
output:

```
pyhton
e
muito
legal

1
2
3
3
2
1
```




````python
def imprime_parametros(*parametros):
    for i in parametros:
        print(i)

imprime_parametros('python','e','muito','legal')

imprime_parametros(1,2,3,3,2,1)
````
