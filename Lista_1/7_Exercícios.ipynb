1) Escreva uma função que conta a quantidade de vogais em um texto  e  armazena  tal  quantidade  em  um  dicionário,  onde  a chave é a vogal considerada.


````python
q_a = q_e = q_i = q_o = q_u = 0
vogais = {"a": q_a, "e": q_e, "i": q_i, "o": q_o, "u": q_u}
texto = str(input("Digite um texto: "))
for i in texto:
    for v in vogais:
        if i == v:
            quant_vogais = 1
            vogais[v] += quant_vogais
        quant_vogais = 0
print(vogais)
````

5) Escreva um programa para armazenar uma agenda de telefones em um dicionário. Cada pessoa pode ter um ou mais telefones e a chave do  dicionário  é  o  nome  da  pessoa.  Seu  programa  deve  ter  as seguintes funções: 

­incluirNovoNome  –  essa  função  acrescenta  um  novo  nome  na agenda,   com   um   ou   mais   telefones.   Ela   deve   receber   como argumentos o nome e os telefones. 

­incluirTelefone  –  essa  função  acrescenta  um  telefone  em  um  nome existente  na  agenda.  Caso  o  nome  não  exista  na  agenda,  você  deve perguntar   se   a   pessoa   deseja incluí­-lo.   Caso   a   resposta   seja afirmativa, use a função anterior para incluir o novo nome.

 excluirTelefone – essa função exclui um telefone de uma pessoa que já está na agenda. Se a pessoa tiver apenas um telefone, ela deve ser excluída da agenda. ­
 
excluirNome – essa função exclui uma pessoa da agenda. ­

consultarTelefone – essa função retorna os telefones de uma pessoa na agenda. 

````python
agenda_de_telefones = {}


def incluirNovoNome(nome, telefones):
    global agenda_de_telefones
    agenda_de_telefones[nome] = telefones


def incluirTelefone(nome, telefone):
    global agenda_de_telefones
    if nome in agenda_de_telefones.keys():
        agenda_de_telefones[nome].append(telefone)
    else:
        resposta = input("deseja incluí­-lo? (s/n")
        if resposta.lower() == "s":
            incluirNovoNome(nome, telefones)


def excluirTelefone(nome, telefone):
    global agenda_de_telefones
    if nome in agenda_de_telefones.keys():
        if len(agenda_de_telefones[nome]) == 1:
            agenda_de_telefones[nome] = None
            print("O usuário %s foi exlcuido" % nome)
        else:
            print(len(agenda_de_telefones[nome]))
            for numero in range(len(agenda_de_telefones[nome]) - 1):
                if telefone == agenda_de_telefones[nome][numero]:
                    agenda_de_telefones[nome].pop(numero)
                    print("O número %s do usuário %s foi excluido" % (telefone, nome))
    else:
        print("Usuário não existe ")


def excluirNome(nome):
    global agenda_de_telefones
    if nome in agenda_de_telefones.keys():
        del agenda_de_telefones[nome]


def consultarTelefone(nome):
    global agenda_de_telefones
    if nome in agenda_de_telefones.keys():
        print(
            "O contato %s possui os seguintes telefones %s"
            % (nome, agenda_de_telefones[nome])
        )
    else:
        print("Contato não existe")


condicao = True
while condicao:
    print("1- ­Adicionar contato na agenda de telefone")
    print("2- Consultar numero do contato")
    print("3- Excluir telefone")
    print("4- Sair \n")
    resposta = int(input("Escolha uma opção: "))

    print("Opção escolhida: ", resposta)
    if resposta == 1:
        nome = input("Digite o nome: ")
        telefones = []
        numero = -1
        while numero != 0:
            numero = int(
                input("Digite o numero de telefone: ou digite zero (0) para parar ")
            )
            if numero != 0:
                telefones.append(numero)
        incluirNovoNome(nome, telefones)
        print("Agenda atual: ", agenda_de_telefones)

    if resposta == 2:
        nome = input("Digite o nome: ")
        consultarTelefone(nome)
        print("Agenda atual: ", agenda_de_telefones)

    if resposta == 3:
        nome = input("Digite o nome: ")
        telefone = int(input("Digite o telefone: "))
        excluirTelefone(nome, telefone)
        print("Agenda atual: ", agenda_de_telefones)

    if resposta == 4:
        condicao = False
        print("Agenda atual: ", agenda_de_telefones)
````
