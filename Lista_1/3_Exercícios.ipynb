1) Ler a temperatura de uma pessoa e exibir a mensagem “Febre Alta” (temp ≥ 39),“Febril" (39 > temp ≥ 37) ou “Sem Febre” (temp <37).

```
temperatura =  float(input("Insira a temperatura: "))
if(temperatura >= 39):
    print("Febre Alta")
elif(39 > temperatura >= 37):
    print("Febril")
elif(temperatura  < 37):
    print("Sem Febre")
```

2) Crie um programa que lê as idades e alturas de alguns alunos. A condição de parada é a altura = 0. Em seguida, o programa deve informar quantos alunos com mais de 18 anos possuem altura inferior à 1.6.

```
ler_dados = True
condicao_alunos = 0    # Condicao >  aluno com mais de 18 anos e altura inferior à 1.6.
while(ler_dados == True):

    idade = int(input("Idade:"))
    altura = float(input("Altura: "))
   
    if( idade > 18  and altura < 1.6):
        condicao_alunos += 1 
    if(altura == 0):
        ler_dados = False

print(condicao_alunos)
```

3) Dado um inteiro não-negativo n, determinar n! Obs.: o n deve ser informado pelo usuário através da função input.

```
n = int(input("Informe o N"))
n_fatorial = 1
for i in range(n, 0, -1):
  n_fatorial = i * n_fatorial
print(n_fatorial)
```

4) Dado uma seqüência de n números inteiros, determinar a soma dos números pares.

Ex.: 

input: n = [1, 2, 4, 5]; 

output:  6 


```
n = [1, 2, 4, 5];
soma = 0
for i in range(len(n)):
    if(n[i] % 2 == 0):
        soma +=  n[i]
print(soma)
```
