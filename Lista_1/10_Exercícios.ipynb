1) Escreva uma classe em Python chamada Retangulo a qual deverá conter os atributos comprimento e largura. 
Implemente os métodos:  

```
calcular a área do retângulo 
calcular a diagonal do rentângulo.
Calcular o perímetro
Alterar valor do comprimento
Alterar valor da largura
```


Faça uma instância da classe e chame os métodos.    

````python
import math

"""
Foi pedido comprimento e largura.
E para calcular a area do retangulo a base é igual ao comprimento e a largura é a altura
"""


class Retangulo:
    def __init__(self, comprimento, largura):
        self.comprimento = comprimento
        self.largura = largura

    def setComprimento(self, comprimento):
        self.comprimento = comprimento

    def getComprimento(self):
        return self.comprimento

    def setLargura(self, largura):
        self.largura = largura

    def getLargura(self):
        return self.largura

    def area_do_retangulo(self):
        return self.largura * self.comprimento

    def diagonal_do_rentangulo(self):
        return round(math.sqrt((self.largura) ** 2 + (self.comprimento) ** 2), 2)

    def perimetro(self):
        return 2 * (self.largura + self.comprimento)


r = Retangulo(10, 5)
print("Largura atual: %s Comprimento atual: %s" % (r.getLargura(), r.getComprimento()))
print("Área do retângulo: ", r.area_do_retangulo())
print("Diagonal do rentângulo: ", r.diagonal_do_rentangulo())
print("Perímetro: ", r.perimetro())

comprimento = int(input("Digite o novo valor do comprimento:"))
largura = int(input("Digite o novo valor de largura:"))

r.setComprimento(comprimento)
r.setLargura(largura)

print("Largura atual: %s Comprimento atual: %s" % (r.getLargura(), r.getComprimento()))
print("Área do retângulo: ", r.area_do_retangulo())
print("Diagonal do rentângulo: ", r.diagonal_do_rentangulo())
print("Perímetro: ", r.perimetro())
````

3) Classe   Funcionário: Implemente a classe Funcionário.  Um  funcionário  tem  um  nome  e  um salário.  Escreva  um  construtor  com  dois  parâmetros (nome  e  salário)  e  o  método aumentarSalario(porcentualDeAumento)  que  aumente  o  salário  do funcionário  em  uma  certa  porcentagem.  Exemplo  de uso:

```
harry=funcionário("Harry",25000) harry.aumentarSalario(10)
```

Faça um programa que teste o método da classe.


````python
class Funcionario:
    """docstring for Funcionario."""

    def __init__(self, nome, salario):
        self.nome = nome
        self.salario = salario

    def setSalario(self, salario):
        self.salario = salario

    def getSalario(self):
        return self.salario

    def aumentarSalario(self, percentualAumento):
        self.salario += (self.salario * percentualAumento) / 100


funcionario = Funcionario("Harry", 25000)
funcionario.aumentarSalario(10)
print("O salário após o aumento é de", funcionario.getSalario())
````

4) Classe Macaco: Desenvolva uma classe Macaco, que possua os atributos nome e bucho (estomago) e pelo menos os métodos comer(), verBucho() e digerir(). Faça um programa ou teste interativamente, criando pelo menos dois macacos, alimentando-os com pelo menos 3 alimentos diferentes e verificando o conteúdo do estomago a cada refeição. Experimente fazer com que um macaco se alimente do outro. É possível criar um macaco canibal? 

````python
class Macaco:
    """docstring for Macaco."""

    def __init__(self, nome):
        self.nome = nome
        self.bucho = []

    def comer(self, comida):
        if comida.lower() != "macaco":
            self.bucho.append(comida)
        else:
            print("Não pode canibalismo")

    def verBucho(self):
        print("No estomago:")
        for i in self.bucho:
            print(i)

    def digerir(self):
        print("Digerindo")
        self.bucho = []


macaco1 = Macaco("Macaco1")
print("Macaco 1: ")
macaco1.comer("Banana")
macaco1.comer("Banana-maçã")
macaco1.verBucho()
macaco1.digerir()
macaco1.verBucho()
print("\n")
print("Macaco 2: ")
macaco2 = Macaco("Macaco2")
macaco2.comer("Banana")
macaco2.comer("Banana-maçã")
macaco2.comer("macaco")
macaco2.verBucho()
````

5) Classe Pessoa: Crie uma classe que modele uma pessoa:

*   Atributos: nome, idade, peso e altura.
*   Métodos: Envelhercer, engordar, emagrecer, crescer. Obs: Por padrão, a cada ano que nossa pessoa envelhece, sendo a idade dela menor que 21 anos, ela deve crescer 0,5 cm.


 


 

````python
class Pessoa:
    """docstring for Pessoa."""

    def __init__(self, nome, idade, peso, altura):
        self.nome = nome
        self.idade = idade
        self.peso = peso
        self.altura = altura

    def envelhecer(self):
        if(self.idade < 21):
            self.altura += 0.5
        self.idade += 1

    def engordar(self, kilo):
        self.peso += kilo

    def emagrecer(self, kilo):
        self.peso -= kilo

    def crescer(self, cres):
        self.altura += cres

    def retornaDados(self):
        print(
            "Nome: %s idade: %s Peso: %s e Altura: %s"
            % (self.nome, self.idade, self.peso, self.altura)
        )


dav = Pessoa("David", 18, 90, 180)
dav.retornaDados() #inicial

dav.crescer(1)
dav.engordar(3)
dav.emagrecer(5)
dav.envelhecer()
dav.retornaDados()
````
