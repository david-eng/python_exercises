1) Desenvolva um programa que solicite a digitação de um número de CPF no formato xxx.xxx.xxx-xx e indique se é um número válido ou inválido através da validação dos dígitos verificadores e dos caracteres de formatação. 

lógica do digito verificador: 
[link text](https://campuscode.com.br/conteudos/o-calculo-do-digito-verificador-do-cpf-e-do-cnpj)



````python
cpf = input("Digite o CPF:")
cpf = cpf.replace(".", "")
cpf = cpf.replace("-", "")

primeiros_nove_digitos = cpf[:9]
primeiros_dez_digitos = cpf[:10]
ultimos_dois_digitos = cpf[9:]

subtrai = 10
subtrai_2 = 11
soma = soma_2 = 0

for i in range(9):
    soma += int(primeiros_nove_digitos[i]) * (subtrai - i)
verifica_primeiro_digito = 11 - (soma % 11)

for i in range(10):
    soma_2 += int(primeiros_dez_digitos[i]) * (subtrai_2 - i)
verifica_segundo_digito = 11 - (soma_2 % 11)

if verifica_primeiro_digito >= 10:
    verifica_primeiro_digito = 0

if verifica_segundo_digito >= 10:
    verifica_segundo_digito = 0


if verifica_primeiro_digito == int(
    ultimos_dois_digitos[0]
) and verifica_segundo_digito == int(ultimos_dois_digitos[1]):
    print("válido")
else:
    print("inválido")
````

2) Faça um programa que solicite a data de nascimento (dd/mm/aaaa) do usuário e imprima a data com o nome do mês por extenso. 


```
    Data de Nascimento: 29/10/1973
    Você nasceu em  29 de Outubro de 1973.
```



````python
meses=('Janeiro','Fevereiro','Março','Abril','Maio','Junho',
           'Julho','Agosto','Setembro','Outubro','Novembro','Dezembro')

data = input("Data de Nascimento:")
data = data.split("/")
dia = int(data[0]);
mes = int(data[1]);
ano = int(data[2]);
print("Você nasceu em  %s de %s de %s." %(dia, meses[mes-1], ano))
````

3) Faça uma função que recebe uma string que representa uma cadeia de DNA e gera a cadeia complementar. A entrada e saída de dados deve ser feita pelo programa principal.

```
Entrada: AATCTGCAC
Saída: TTAGACGTG
```



````python

def representa_cadeia_dna(dna):
    res = str("")
    for i in dna:
        if(i == "A"):
            res = res + "T"
        if(i == "T"):
            res = res + "A"
        if(i == "C"):
            res = res + "G"
        if(i == "G"):
            res = res + "C"
    return(res)

dna = input("Entrada: ")
print("Saída : %s" % representa_cadeia_dna(dna))
````

4) Escreva um programa em Python que conte o número de ocorrência de cada palavra em uma dada sentença.

````python
sentenca = "Uma vez Flamengo, sempre Flamengo Flamengo sempre eu hei de ser"
sentenca = sentenca.replace(",", "")
palavras = sentenca.split(" ")
palavras = sorted(palavras)
conta_iguais = 0
ocorrencia_palavra = []
for i in range(len(palavras)):
    for x in range(len(palavras)):
        if palavras[i] == palavras[x]:
            conta_iguais += 1

    ocorrencia_palavra.append(conta_iguais)
    if conta_iguais > 1 and palavras[i] == palavras[i - 1]:
        pass
    else:
        print(
            "A palavra '%s'  apareceu %s vez(es) "
            % (palavras[i], ocorrencia_palavra[i])
        )
    conta_iguais = 0
````
